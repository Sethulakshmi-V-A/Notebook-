####  3. The manager wants to find out whether the same amount was spent for the three advertisements (TV , Radio and Newspaper).     Comment on your findings.

In order to check whether same amount was spent on the advertisements we can compare the equality of means of the three advertisements.This can be done using the ANOVA.

Consider the hypothesis

H0 : Same amounts were spent for three advertisements
H1 : Different amounts were spent for the advertisements   

```python
from scipy.stats import f_oneway 
```

```python
f_val,p_val=f_oneway(adv_data['TV'],adv_data['Radio'],adv_data['Newspaper'])
print('F value:',f_val,'\nP value:',p_val)
```

```python
alpha=0.05
if p_val>= alpha:
    print('Sample fails to reject null hypotheses.','\n Same amounts were spent for the three advertisements.')
else:
     print('Sample fails to accept null hypothesis.','\n Different amounts were spent for the three advertisements.')     
```
